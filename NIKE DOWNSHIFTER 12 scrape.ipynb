```python
import numpy as np
import pandas as pd
import seaborn
from bs4 import BeautifulSoup
import requests
```

```python
#negative reviews
nrating=[]
nreview=[]
for i in range(1,80):
    #url='https://www.flipkart.com/nike-downshifter-12-running-shoes-men/product-reviews/itm229e6f70c1aed?pid=SHOGKV3JGHBD5UYG&lid=LSTSHOGKV3JGHBD5UYGE23IGO&aid=overall&certifiedBuyer=false&sortOrder=MOST_RECENT&page='+str(i)
    url='https://www.flipkart.com/nike-downshifter-12-running-shoes-men/product-reviews/itm229e6f70c1aed?pid=SHOGKV3JGHBD5UYG&lid=LSTSHOGKV3JGHBD5UYGE23IGO&aid=overall&certifiedBuyer=false&sortOrder=NEGATIVE_FIRST'
    response = requests.get(url)
    htmlcontent = response.content
    response = requests.get(url)
    htmlcontent = response.content
    soup = BeautifulSoup(htmlcontent,"html.parser")
    nrat1=soup.find_all('div',attrs={'class':'_3LWZlK _1rdVr6 _1BLPMq _3B8WaH'})
    nrat2=soup.find_all('div',attrs={'class':'_3LWZlK _32lA32 _1BLPMq _3B8WaH'})
    nrev=(soup.find_all('div',attrs={'class':'_6K-7Co'}))
    for i in nrev:
        nreview.append(i.text)
    for i in nrat1:
        nrating.append(i.text)
    for i in nrat2:
        nrating.append(i.text)
print(nreview)
print(nrating)
```

```python
len(nrating),len(nreview),type(nrating),type(nreview)
```

```python
#positive reviews
prating=[]
preview=[]
for i in range(1,80):
    url='https://www.flipkart.com/nike-downshifter-12-running-shoes-men/product-reviews/itm229e6f70c1aed?pid=SHOGKV3JGHBD5UYG&lid=LSTSHOGKV3JGHBD5UYGE23IGO&aid=overall&certifiedBuyer=false&sortOrder=POSITIVE_FIRST'
    response = requests.get(url)
    htmlcontent = response.content
    response = requests.get(url)
    htmlcontent = response.content
    soup = BeautifulSoup(htmlcontent,"html.parser")
    prev=(soup.find_all('div',attrs={'class':'_6K-7Co'}))
    prat=soup.find_all('div',attrs={'class':'_3LWZlK _1BLPMq _3B8WaH'})
    for i in prev:
        preview.append(i.text)
    for i in prat:
        prating.append(i.text)
print(preview)
print(prating)
```

```python
len(prating),len(preview),type(prating),type(preview) #100 990,120 1190
```

```python
Review=nreview+preview
Review
```

```python
Rating=nrating+prating
Rating
```

```python
print("Number of records scraped : ",len(Review))
print("Number of records scraped : ",len(Rating))
```

```python
# converting our reviews list to dataframe
df = pd.DataFrame(Review,
                  columns =['review'])
```

```python
df["rating"]=Rating
```
